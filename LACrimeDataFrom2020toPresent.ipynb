Will also look at the data

```python
data = pd.read_csv('Crime_Data_from_2020_to_Present.csv')
data.head()
```

```python
data.info()
```

```python
data.isnull().mean() # will look at what percentage of the data is empty. 
```

```python
df = data.rename(columns={'Date Rptd':'DateRptd'})
df.head()
```

```python
DR =df.DateRptd.str.split(expand=True)
DR.head(20)
```

```python
df['DateRptd'] = DR[0].values
df.head()
```

```python
DOC =df["DATE OCC"].str.split(expand=True)
DOC.head()
df["DATE OCC"] = DOC[0].values
df.head()
```

```python
df.isnull().mean() # will look at what percentage of the data is empty. 
```

Will be dropping Columns for MOCODES, Crm Cd 1, Crm Cd 2, Crm Cd 3, Crm Cd 4, and Cross Street. These are not believed to be necessary as they do not add any value to the analysis. 

```python
DataDrop = df
```

```python
DD = DataDrop.drop(['Mocodes', 'Crm Cd 1' , 'Crm Cd 2' , 'Crm Cd 3' , 'Crm Cd 4', 'Cross Street'], axis=1)
```

```python
DD.columns
```

```python
DD.isnull().mean()
```

Will fill in the missing valies for the Vict Sex, Vict Descent, Premis Cd, Premis Desc, Weapon Used Cd, and Weapon Desc as null. Do not want to drop null values as data is usefull. 

```python
DFNAN = DD.fillna('NAN')
```
